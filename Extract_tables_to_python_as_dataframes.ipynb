```python
import os
import pandas as pd
import zipfile

# 1. Apply your temporary patch to ignore CRC checksum errors
def ignore_crc(self, newdata):
    if self._eof and self._running_crc != self._expected_crc:
        pass  # Bypass metadata errors quietly

zipfile.ZipExtFile._update_crc = ignore_crc

# 2. File Path
file_path = r"enter_file_path"

try:
    # 3. Read the sheet names dynamically
    excel_reader = pd.ExcelFile(file_path)
    all_table_names = excel_reader.sheet_names
    
    print(f"📦 Found {len(all_table_names)} tables inside the workbook.\n")
    
    # 4. Extract data and inject them straight into the global namespace
    for name in all_table_names:
        print(f"   ⏳ Loading and creating standalone variable: {name}")
        
        # Read the sheet data
        df = pd.read_excel(file_path, sheet_name=name)
        
        # MAGIC LINE: This registers the string name as a direct variable in your notebook
        globals()[name] = df
        
    print(f"\n🎉 Success! All tables have been converted into direct variables.")
    print(f"👉 You can now call them natively: {', '.join(all_table_names)}\n")

except Exception as e:
    print(f"\n❌ Error converting tables: {e}")
```
